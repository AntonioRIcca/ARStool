```python
import pandas as pd
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# Underground network
RSADI_und_jan = pd.read_excel('R_und_Jan.xlsx', sheet_name='RSADI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSADI_und_may = pd.read_excel('R_und_May.xlsx', sheet_name='RSADI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSADI_und_aug = pd.read_excel('R_und_Aug.xlsx', sheet_name='RSADI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSADI_und_nov = pd.read_excel('R_und_Nov.xlsx', sheet_name='RSADI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)

RSLSI_und_jan = pd.read_excel('R_und_Jan.xlsx', sheet_name='RSLSI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSLSI_und_may = pd.read_excel('R_und_May.xlsx', sheet_name='RSLSI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSLSI_und_aug = pd.read_excel('R_und_Aug.xlsx', sheet_name='RSLSI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSLSI_und_nov = pd.read_excel('R_und_Nov.xlsx', sheet_name='RSLSI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)

RSADI_und = (RSADI_und_jan + RSADI_und_may + RSADI_und_aug + RSADI_und_nov) / 4
RSADI_und = RSADI_und.to_numpy()

RSLSI_und = (RSLSI_und_jan + RSLSI_und_may + RSLSI_und_aug + RSLSI_und_nov) / 4
RSLSI_und = RSLSI_und.to_numpy()

# EV PARKING network
RSADI_ev_jan = pd.read_excel('R_ev_Jan.xlsx', sheet_name='RSADI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSADI_ev_may = pd.read_excel('R_ev_May.xlsx', sheet_name='RSADI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSADI_ev_aug = pd.read_excel('R_ev_Aug.xlsx', sheet_name='RSADI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSADI_ev_nov = pd.read_excel('R_ev_Nov.xlsx', sheet_name='RSADI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)

RSLSI_ev_jan = pd.read_excel('R_ev_Jan.xlsx', sheet_name='RSLSI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSLSI_ev_may = pd.read_excel('R_ev_May.xlsx', sheet_name='RSLSI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSLSI_ev_aug = pd.read_excel('R_ev_Aug.xlsx', sheet_name='RSLSI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)
RSLSI_ev_nov = pd.read_excel('R_ev_Nov.xlsx', sheet_name='RSLSI', usecols="B:D", skiprows=1, nrows=95, 
                              header=None)

RSADI_ev = (RSADI_ev_jan + RSADI_ev_may + RSADI_ev_aug + RSADI_ev_nov) / 4
RSADI_ev = RSADI_ev.to_numpy()

RSLSI_ev = (RSLSI_ev_jan + RSLSI_ev_may + RSLSI_ev_aug + RSLSI_ev_nov) / 4
RSLSI_ev = RSLSI_ev.to_numpy()


# RES network
RSADI_res_jan = pd.read_excel('R_res_Jan.xlsx', sheet_name='RSADI', usecols="B:F", skiprows=1, nrows=95, 
                              header=None)
RSADI_res_may = pd.read_excel('R_res_May.xlsx', sheet_name='RSADI', usecols="B:F", skiprows=1, nrows=95, 
                              header=None)
RSADI_res_aug = pd.read_excel('R_res_Aug.xlsx', sheet_name='RSADI', usecols="B:F", skiprows=1, nrows=95, 
                              header=None)
RSADI_res_nov = pd.read_excel('R_res_Nov.xlsx', sheet_name='RSADI', usecols="B:F", skiprows=1, nrows=95, 
                              header=None)

RSLSI_res_jan = pd.read_excel('R_res_Jan.xlsx', sheet_name='RSLSI', usecols="B:F", skiprows=1, nrows=95, 
                              header=None)
RSLSI_res_may = pd.read_excel('R_res_May.xlsx', sheet_name='RSLSI', usecols="B:F", skiprows=1, nrows=95, 
                              header=None)
RSLSI_res_aug = pd.read_excel('R_res_Aug.xlsx', sheet_name='RSLSI', usecols="B:F", skiprows=1, nrows=95, 
                              header=None)
RSLSI_res_nov = pd.read_excel('R_res_Nov.xlsx', sheet_name='RSLSI', usecols="B:F", skiprows=1, nrows=95, 
                              header=None)

RSADI_res = (RSADI_res_jan + RSADI_res_may + RSADI_res_aug + RSADI_res_nov) / 4
RSADI_res = RSADI_res.to_numpy()

RSLSI_res = (RSLSI_res_jan + RSLSI_res_may + RSLSI_res_aug + RSLSI_res_nov) / 4
RSLSI_res = RSLSI_res.to_numpy()
```
